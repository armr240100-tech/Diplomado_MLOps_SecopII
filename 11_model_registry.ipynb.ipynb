```python
import mlflow
from mlflow.tracking import MlflowClient

# 1. Conexión (usando la dirección que ya vimos que funciona)
mlflow.set_tracking_uri("http://mlflow:5000")

# 2. Registrar el modelo en el catálogo central
# Esto le da un nombre oficial al modelo para que otros lo encuentren
model_name = "Modelo_Predictor_SECOP_II"
run_id = mlflow.search_runs(experiment_names=["Proyecto_SECOP_II"]).iloc[0].run_id

model_uri = f"runs:/{run_id}/model"
mlflow.register_model(model_uri, model_name)

# 3. PROMOCIÓN A PRODUCCIÓN
# Aquí le ponemos la etiqueta oficial de 'Production'
client = MlflowClient()
client.transition_model_version_stage(
    name=model_name,
    version=1,
    stage="Production"
)

print(f"El modelo '{model_name}' ha sido registrado y promovido a PRODUCCIÓN.")
```
